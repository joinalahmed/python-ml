## Hypothesis Testing using Kolmogorov–Smirnov Test

```python
def ks_test(dist):
    """Kolmogorov-Smirnov Test where null hypothesis (H0) is dist follows uniform distribution, 
       with alternative (H1) that dist is not uniformly distributed.
    """

    from scipy import stats

    _min, _max, mu, sigma = dist.min(), dist.max(), dist.mean(), dist.std()
    print('KS Test against {0}...'.format(dist.name))
    D, p = stats.kstest(dist, 'uniform', args=(_min, _max-_min), alternative='two-sided')
    print('H0: uniform distribution.  D-statistic: {:.6f}, p: {:.4%}'.format(D, p))
    D, p = stats.kstest(dist, 'norm', args=(mu, sigma), alternative='two-sided')
    print('H0: gaussian distribution. D-statistic: {:.6f}, p: {:.4%}'.format(D, p))
```

```python
ks_test(raw['Temperature'])
```

since p > 0.05, we cannot reject null hypothesis (uniform) and therefore accept that it's an uniform distribution.

```python
ks_test(raw['Flow'])
```

since p > 0.05, we cannot reject null hypothesis (uniform) and therefore accept that it's an uniform distribution.

## Model Selection

```python
X = raw[['Flow', 'Temperature']]
y = raw['Power']
```

```python
model_candidates = [
    linear_model.LinearRegression(),
    linear_model.RidgeCV(),
    linear_model.LassoCV(),
    linear_model.ElasticNetCV(),
    tree.DecisionTreeRegressor(),
    ensemble.RandomForestRegressor(),
    svm.SVR()
]
```

```python
kf = model_selection.KFold(n_splits=5)

for reg in model_candidates:
    composed_reg = pipeline.make_pipeline(preprocessing.StandardScaler(), reg)
    kf_mse = model_selection.cross_val_score(composed_reg, X, y, scoring='neg_mean_squared_error', cv=kf)
    kf_r2 = model_selection.cross_val_score(composed_reg, X, y, scoring='r2', cv=kf)
    print('{0:<25}: R-squared: {1:.6f};   Mean Squared Error: {2:.6f}'.format(
                                                            reg.__class__.__name__,
                                                            kf_r2.mean(),
                                                            abs(kf_mse.mean())
        ))
```

## Linear Regression

```python
reg = linear_model.LinearRegression()
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=.2, random_state=1)
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
X_train_std = scaler.transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
reg.fit(X_train_std, y_train)
```

```python
y_pred = reg.predict(X_test_std)
```

```python
print('{0:<25}: R-squared: {1:.6f};   Mean Squared Error: {2:.6f}'.format(
    reg.__class__.__name__, 
    metrics.r2_score(y_test, y_pred),
    metrics.mean_squared_error(y_test, y_pred)
    ))
```

```python
residuals = y_pred - y_test
residuals.name = 'residuals'
```

```python
ks_test(residuals)
sns.distplot(residuals)
```

since p > 0.05, we cannot reject null hypothesis (gaussian) and accept that residuals follow gaussian distribution.
